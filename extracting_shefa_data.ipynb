```python
import csv
from bs4 import BeautifulSoup
import requests
```

```python
def get_data(n):
    target_pages = []
    with open('D:/data_analysis/webscraping/shefa-sa.com/data.csv','r',encoding='utf-8') as file:
        rows = csv.reader(file)
        for i in range(n):    
            target_pages.append(next(rows))
    return target_pages
```

```python
def extract_data(n):
    target = get_data(n)
    extracted_cases = []
    extracted_data = []
    for page in target:
        if len(page)>2:
            soup = BeautifulSoup(page[2])
            cases = soup.find_all(attrs={'class':'col-lg-4'})
            for case in cases:
                c_id = case.find(attrs={'class':'case-id'}).text
                spec = case.find(attrs={'class':'ml-2'}).text
                loc = case.find(attrs={'class':'cases-header'}).text
                desc = case.find(attrs={'class':'condition-description'}).text
                extracted_cases.append(case)
                extracted_data.append((c_id,spec,loc,desc))
    return extracted_cases,extracted_data
```
